```python
import numpy as np # Utilisé pour les opérations numériques avancées sur les tableaux (arrays).
import pandas as pd # Utilisé pour la manipulation et l'analyse de données, notamment avec les DataFrames.
import matplotlib.pyplot as plt # Utilisé pour créer des visualisations de données statiques, comme des graphiques en lignes, des histogrammes, etc.
import plotly.express as px # Utilisé pour créer des visualisations interactives et dynamiques.
import seaborn as sns # Basé sur matplotlib, utilisé pour créer des visualisations statistiques plus sophistiquées.
```

```python
# Sélectionne uniquement les colonnes de types numériques (int64 et float64) du DataFrame 'df' et les assigne à un nouveau DataFrame 'df_numeric'
df_numeric = df.select_dtypes(include=['int64', 'float64'])
```

```python
# Crée une nouvelle figure de taille 20x18 pouces et affiche une carte thermique des corrélations entre les colonnes numériques du DataFrame 'df_numeric'
plt.figure(figsize=(20,18))
sns.heatmap(df_numeric.corr(), annot=True,linewidths=.5, cmap="Purples")
```

```python
# Crée une matrice de graphes de dispersion (scatter plots) pour chaque paire de colonnes numériques dans 'df_numeric' et colorie les points en fonction de la colonne 'Vues'
sns.pairplot(data=df_numeric, hue='Vues')
plt.show()
```

```python
# Affiche des statistiques descriptives pour chaque colonne numérique du DataFrame 'df', incluant la moyenne, l'écart-type, les valeurs min et max, et les quartiles
df.describe()
```

```python
# Données
variables = ['Vues', 'Utilisateurs', 'Vues par utilisateur', 'Durée d\'engagement moyenne', 'Nombre d\'événements']
count = [5000, 5000, 5000, 5000, 5000]
mean = [193.293400, 118.320000, 1.305475, 23.987946, 546.909600]
std = [2674.425852, 994.637126, 0.751526, 17.311588, 7689.707459]
min_value = [26, 1, 0.878788, 0.333333, 34]
max_value = [162073, 48895, 32, 171.814897, 465066]

# Création des graphiques
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Diagrammes à barres pour les statistiques de base
for i in range(len(variables)):
    row = i // 3
    col = i % 3
    ax = axs[row, col]
    ax.bar(['Count', 'Mean', 'Std', 'Min', 'Max'], [count[i], mean[i], std[i], min_value[i], max_value[i]])
    ax.set_title(variables[i])

plt.tight_layout()
plt.show()
```
